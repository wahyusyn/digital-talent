```python
#koneksi database
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
#fungsi untuk query dijalankan ditampilkan
def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Prepare a list of offices sorted by country, state, city.
run("SELECT country,state,city FROM offices ORDER BY country,state,city")
```

```python
#2. How many employees are there in the company?
run("SELECT COUNT(employeeNumber) AS pegawai,(\"22 orang + 1 presiden\") AS keterangan FROM employees")
```

```python
#3. What is the total of payments received?
run("SELECT SUM(amount) AS totalPembayaran FROM payments")
```

```python
#4. List the product lines that contain 'Cars'.
run("SELECT productLine FROM productlines WHERE productLine LIKE '%cars%'")
```

```python
#5. Report total payments for October 28, 2004
run("SELECT amount FROM payments WHERE paymentDate='2004-10-28'")
```

```python
#6. Report those payments greater than $100,000
run("SELECT customerNumber,checkNumber,paymentDate,amount FROM payments WHERE amount>'100000'")
```

```python
#7. List the products in each product line
run("SELECT productLine,productName,quantityInStock,buyPrice,MSRP FROM products ORDER BY productLine")
```

```python
#8. How many products in each product line?
run("SELECT productLine,COUNT(productCode) AS jumlah FROM products GROUP BY productLine ORDER BY jumlah DESC")
```

```python
#9. What is the minimum payment received?
run("SELECT MIN(amount) AS minumum FROM payments")
```

```python
#10. List all payments greater than twice the average payment
rataX2=run("SELECT AVG(amount)*2 rataRata FROM payments")
rataX2=rataX2['rataRata'][0]
print("2 Kali Rata-Rata =",rataX2)
run(f"SELECT amount FROM payments WHERE amount>'{rataX2}' ORDER BY amount DESC")
```

```python
#11. What is the average percentage markup of the MSRP on buyPrice?
dataMark=run("""SELECT productCode,buyPrice,MSRP,((MSRP-buyPrice)/buyPrice*100) AS markup
    FROM products
    ORDER BY markup DESC
    LIMIT 7""")
print("7 Markup teratas :\n",dataMark,"\n")
rataMark=run("SELECT AVG(((MSRP-buyPrice)/buyPrice*100)) AS rataMarkUp FROM products")
print("Rata-Rata Persentase Markup :",rataMark['rataMarkUp'][0],"%")
```

```python
#12. How many distinct products does ClassicModels sell?
produkPenjualan=run("SELECT COUNT(DISTINCT(orderdetails.productCode)) uniqProduk FROM orderdetails")
print("Jumlah uniq produk terjual :",produkPenjualan['uniqProduk'][0])
```

```python
#13. Report the name and city of customers who don't have sales representatives?
run("SELECT customerName,city FROM customers WHERE salesRepEmployeeNumber IS NULL")
```

```python
#14. What are the names of executives with VP or Manager in their title?
#Use the CONCAT function to combine the employee's first name and last name into a single field for reporting
run("""SELECT CONCAT(firstName,' ',lastName) AS name,jobTitle
    FROM employees
    WHERE jobTitle LIKE '%VP%' OR jobTitle LIKE '%Manager%'""")
```

```python
#15. Which orders have a value greater than $5,000?
run("""SELECT orderNumber,SUM(quantityOrdered*priceEach) AS nilai
    FROM orderdetails
    GROUP BY orderNumber
    HAVING nilai>'5000'
    ORDER BY nilai DESC""")
```
