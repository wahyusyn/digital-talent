```python
#koneksi database
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')
#fungsi untuk query dijalankan ditampilkan
def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## 1 to Many

```python
#1. Report the account representative for each customer
run("""SELECT customers.customerName,
    CONCAT(employees.firstName,' ',employees.lastName) AS accRepresentative
FROM customers
LEFT JOIN employees ON customers.salesRepEmployeeNumber = employees.employeeNumber
ORDER BY employees.firstName DESC
""")
```

```python
#2. Report total payments for Atelier graphique
run("""SELECT SUM(payments.amount) AS total
FROM payments
INNER JOIN customers ON payments.customerNumber = customers.customerNumber
WHERE customers.customerName = 'Atelier graphique'
""")
```

```python
#3. Report the total payments by date
run("SELECT paymentDate,SUM(amount) AS total FROM payments GROUP BY paymentDate ORDER BY total DESC")
```

```python
#4. Report the products that have not been sold
run("""SELECT products.productCode,products.productName,products.productLine,
    products.quantityInStock,products.buyPrice,products.MSRP
FROM products
LEFT JOIN orderdetails ON orderdetails.productCode = products.productCode
WHERE orderdetails.orderNumber IS NULL
""")
```

```python
#5. List the amount paid by each customer
run("""SELECT customers.customerNumber,customers.customerName,Sum(payments.amount) AS totalAmount
FROM customers
INNER JOIN payments ON payments.customerNumber = customers.customerNumber
GROUP BY customers.customerNumber
ORDER BY totalAmount DESC
""")
```

```python
#6. How many orders have been placed by Herkku Gifts?
run("""SELECT COUNT(orders.orderNumber) AS jumlah
FROM orders
INNER JOIN customers ON orders.customerNumber = customers.customerNumber
WHERE customers.customerName LIKE 'herkku%'
""")
```

```python
#7. Who are the employees in Boston?
run("""SELECT employees.employeeNumber,
    CONCAT(employees.firstName,' ',employees.lastName) AS employeeName
FROM employees
INNER JOIN offices ON employees.officeCode = offices.officeCode
WHERE offices.city = 'boston'
""")
```

```python
#8. Report those payments greater than $100,000. Sort the report so the customer who made the highest payment appears first
run("""SELECT customers.customerNumber,customers.customerName,Sum(payments.amount) AS totalAmount
FROM customers
INNER JOIN payments ON payments.customerNumber = customers.customerNumber
GROUP BY customers.customerNumber
HAVING totalAmount > '100000'
ORDER BY totalAmount DESC
""")
```

```python
#9. List the value of 'On Hold' orders
run("""SELECT orderdetails.orderNumber,
    SUM(orderdetails.quantityOrdered*orderdetails.priceEach) AS nilai
FROM orderdetails
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE orders.`status` = 'On Hold'
GROUP BY orderdetails.orderNumber
ORDER BY nilai DESC
""")
```

```python
#10. Report the number of orders 'On Hold' for each customer
run("""SELECT customers.customerName,
    Sum(orderdetails.quantityOrdered) AS jumlahPieceOrder
FROM orderdetails
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
INNER JOIN customers ON orders.customerNumber = customers.customerNumber
WHERE orders.`status` = 'On Hold'
GROUP BY customers.customerName
""")
```

## Many to Many

```python
#1. List products sold by order date
run("""SELECT products.productCode,products.productName,orders.orderDate
FROM products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE orders.`status` NOT IN ('On Hold', 'Disputed')
GROUP BY products.productCode
ORDER BY orders.orderDate ASC
""")
```

```python
#2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck
run("""SELECT orders.orderNumber,orders.orderDate
FROM orders
INNER JOIN orderdetails ON orderdetails.orderNumber = orders.orderNumber
INNER JOIN products ON orderdetails.productCode = products.productCode
WHERE products.productName = '1940 Ford Pickup Truck'
ORDER BY orders.orderDate DESC
""")
```

```python
#3. List the names of customers and their corresponding order number where
#   a particular order from that customer has a value greater than $25,000?
run("""SELECT customers.customerName,orders.orderNumber,
    SUM(orderdetails.quantityOrdered*orderdetails.priceEach) AS nilai
FROM customers
INNER JOIN orders ON orders.customerNumber = customers.customerNumber
INNER JOIN orderdetails ON orderdetails.orderNumber = orders.orderNumber
GROUP BY orders.orderNumber
HAVING nilai>'25000'
ORDER BY customers.customerName ASC
""")
```

```python
#4. Are there any products that appear on all orders?
print("Dicari produk yang orderNumbernya 'NULL', kalau ada artinya 'Tidak Ada Satupun Produk yang Muncul di Setiap Order'")
run("""SELECT products.productCode
FROM products
LEFT JOIN orderdetails ON orderdetails.productCode = products.productCode
WHERE orderdetails.orderNumber IS NULL
""")
```

```python
#5. List the names of products sold at less than 80% of the MSRP
run("""SELECT products.productName,products.MSRP,
    orderdetails.priceEach,(products.MSRP*80/100) AS `80markup`
FROM products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
HAVING orderdetails.priceEach<`80markup`
""")
```

```python
#6. Reports those products that have been sold with a markup of 100% or more
#   (i.e.,  the priceEach is at least twice the buyPrice)
run("""SELECT products.productName,products.buyPrice,orderdetails.priceEach
FROM products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
GROUP BY products.productCode
HAVING (orderdetails.priceEach-products.buyPrice)/products.buyPrice*100 >= '100'
""")
```

```python
#7. List the products ordered on a Monday
run("""SELECT products.productName,Count(orderdetails.orderNumber) AS jumlahOrderDiSenin
FROM products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE DAYNAME(orders.orderDate) = 'Monday'
GROUP BY products.productCode
ORDER BY products.productCode ASC
""")
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
data=run("""SELECT products.productCode AS kode, products.productName,
    products.quantityInStock AS stokAwal,
    (SELECT stokAwal-SUM(orderdetails.quantityOrdered)
    FROM orderdetails INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
    WHERE orderdetails.productCode = kode AND orders.`status`='Shipped') AS sisaStok
FROM products
INNER JOIN orderdetails ON orderdetails.productCode = products.productCode
INNER JOIN orders ON orderdetails.orderNumber = orders.orderNumber
WHERE orders.`status` = 'On Hold'
""")
jumlahOnHandStok=(data['stokAwal'].sum()-data['sisaStok'].sum())
print("Jumlah On Hand Stok :",jumlahOnHandStok)
```
