## 2. Statistik Data

## 2.1 Statistik Deskriptif data_mel

### 2.1.1 Data Per Liga dari Musim 2013/2014 sampai 2016/2017:

```python
#Tampilkan Data
data_mel.loc[(data_mel["Season"]>="2013/2014") & (data_mel["Season"]<="2016/2017")].groupby("League").sum()
```

```python
#Jumlahkan Data Per Liga dari Musim 2013/2014 sampai 2016/2017
data_melPerLiga=data_mel.loc[(data_mel["Season"]>="2013/2014") & (data_mel["Season"]<="2016/2017")].groupby("League").sum()
```

```python
#Deskripsi Statistik Gol Liga
dataGol=data_melPerLiga.describe()["Goals"]
print(dataGol)
```

```python
jumlahLiga=int(dataGol["count"])
rataGol=dataGol["mean"]
#stdGol=dataGol["std"]
minGol=int(dataGol["min"])
maxGol=int(dataGol["max"])
'''rowMinGol=data_melPerLiga.loc[(data_melPerLiga["Goals"]==minGol)]
w=rowMinGol.iloc[0:1,0:0]'''
print("\n # Statistik Gol Per Liga #\n")
print("Jumlah Liga\t\t:",jumlahLiga)
print("Rata-Rata Gol\t\t:",rataGol)
#print("Standar Deviasi\t\t:",stdGol)
print("Gol Paling Sedikit\t:",minGol)
print("Gol Paling Banyak\t:",maxGol)
```

```python
#Deskripsi Statistik Tendangan/Tembakan Ke Arah Lawan
dataShot=data_melPerLiga.describe()["Shots"]
print(dataShot)
```

```python
jumlahLiga=int(dataShot["count"])
rataShot=dataShot["mean"]
minShot=int(dataShot["min"])
maxShot=int(dataShot["max"])
print("\n # Statistik Tembakan Per Liga #\n")
print("Jumlah Liga\t\t:",jumlahLiga)
print("Rata-Rata Tembakan\t:",rataShot)
print("Tembakan Paling Sedikit\t:",minShot)
print("Tembakan Paling Banyak\t:",maxShot)
```

### 2.1.2 Data Per Klub (EPL) dari Musim 2013/2014 sampai 2016/2017:

```python
#Tampilkan Data
data_mel.loc[(data_mel["Season"]>="2013/2014") & (data_mel["Season"]<="2016/2017") & (data_mel["League"]=="English Premier League")].groupby("Team").sum()
```

```python
#Jumlahkan Data Per Klub RPL dari Musim 2013/1024 sampai 2016/2017
data_melPerKlubEpl=data_mel.loc[(data_mel["Season"]>="2013/2014") & (data_mel["Season"]<="2016/2017") & (data_mel["League"]=="English Premier League")].groupby("Team").sum()
```

```python
#Deskripsi Statistik Kartu Kuning
dataYellow=data_melPerKlubEpl.describe()["Yellow_Cards"]
print(dataYellow)
```

```python
jumlahKlub=int(dataYellow["count"])
rataYellow=dataYellow["mean"]
minYellow=int(dataYellow["min"])
maxYellow=int(dataYellow["max"])
print("\n # Statistik Kartu Kuning EPL Per Klub #\n")
print("Jumlah Klub\t\t\t:",jumlahKlub)
print("Rata-Rata Kartu Kuning\t\t:",rataYellow)
print("Kartu Kuning Paling Sedikit\t:",minYellow)
print("Kartu Kuning Paling Banyak\t:",maxYellow)
```

## 2.2 Statistik Deskriptif data_epl

Data Liga Primer Ingris dari Musim 2013/1024 sampai 2016/2017:

```python
#Tampilkan Data
data_epl.loc[(data_epl["YEAR"]>="2013/2014") & (data_epl["YEAR"]<="2016/2017")].groupby("Team").sum()
```

```python
#Data Per Klub Liga Primer Inggris Musim 2013/2014 sampai 2016/2017
data_eplPerKlub=data_epl.loc[(data_epl["YEAR"]>="2013/2014") & (data_epl["YEAR"]<="2016/2017")].groupby("Team").sum()
```

```python
#Deskripsi Statistik Gol Klub
dataGolEpl=data_eplPerKlub.describe()["F"]
print(dataGolEpl)
```

```python
jumlahKlub=int(dataGolEpl["count"])
rataGolEpl=dataGolEpl["mean"]
minGolEpl=int(dataGolEpl["min"])
maxGolEpl=int(dataGolEpl["max"])
print("\n # Statistik Gol Per Klub EPL#\n")
print("Jumlah Klub\t\t:",jumlahKlub)
print("Rata-Rata Gol\t\t:",rataGolEpl)
print("Gol Paling Sedikit\t:",minGolEpl)
print("Gol Paling Banyak\t:",maxGolEpl)
```

```python
#Deskripsi Statistik Poin Klub
dataPoin=data_eplPerKlub.describe()["P"]
print(dataPoin)
```

```python
jumlahKlub=int(dataPoin["count"])
rataPoin=dataPoin["mean"]
minPoin=int(dataPoin["min"])
maxPoin=int(dataPoin["max"])
print("\n # Statistik Poin Per Klub EPL#\n")
print("Jumlah Klub\t\t:",jumlahKlub)
print("Rata-Rata Poin\t\t:",rataPoin)
print("Poin Paling Sedikit\t:",minPoin)
print("Poin Paling Banyak\t:",maxPoin)
```

## 3. Visualisasi Data

### 3.1 Grafik Liga Eropa dari Musim 2013/2014 sampai 2016/2017:

```python
ax=data_mel.loc[(data_mel["Season"]>="2013/2014") & (data_mel["Season"]<="2016/2017")].groupby("League").sum()[["Goals","Shots_On_Goal","Yellow_Cards"]]
ax=ax.plot(kind="bar",title="Jumlah Gol, Tembakan Terarah, dan Kartu Kuning Berdasarkan Liga",figsize=(9,5),legend=True,fontsize=11)
ax.set_xlabel("Liga",fontsize=13)
ax.set_ylabel("Jumlah",fontsize=13)
```

### 3.2 Grafik EPL dari Musim 2013/2014 sampai 2016/2017:

```python
ax=data_epl.loc[(data_epl["YEAR"]>="2013/2014") & (data_epl["YEAR"]<="2016/2017")].groupby("Team").sum()[["MP","W","F","P"]]
ax=ax.plot(kind="bar",title="Jumlah Game, Kemenangan, Gol, dan Poin Per Klub EPL",figsize=(17,7),legend=True,fontsize=11)
ax.set_xlabel("KLub",fontsize=13)
ax.set_ylabel("Gol",fontsize=13)
```
