Diberikan sekumpulan Nilai Random antara 0 sampai dengan 100 sebanyak 250 nilai. Buatlah program untuk melakukan hal berikut:
1. Tampilkan Angka Ganjil dari ke 50 nilai tersebut
2. Tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. Tampilkan standar deviasi dari sekumpulan nilai tersebut
4. Tampilkan nilai maksimum dan minimum
5. Tampilkan modus dari sekumpulan nilai tersebut

Berikut Kodenya:

```python
#tugas 2 wahyu; saved 2018-10-22 23:26:53
import random

# GENERATE RANDOM
banyakNilai=250
nilaiTugas=[]
for i in range(banyakNilai):
    nilaiTugas.append(random.randrange(1,100,1))

nilaiGanjil=[]#untuk nilai gangil
besarDari70=[]#untuk nilai > 70
nilaiTugasKuadrat=[]#untuk std deviasi

for i in nilaiTugas:
    # BILANGAN GANJIL
    if nilaiTugas[i] % 2 != 0:
        nilaiGanjil.append(nilaiTugas[i])
    # NILAI > 70
    if nilaiTugas[i]>70:
        besarDari70.append(nilaiTugas[i])
    nilaiTugasKuadrat.append(pow(nilaiTugas[i],2))#untuk std deviasi

# STANDAR DEVIASI (manual)
jumlahNilaiTugas=sum(nilaiTugas)
jumlahNilaiTugasKuadrat=sum(nilaiTugasKuadrat)
kuadratJumlahNilaiTugas=pow(jumlahNilaiTugas,2)
varian=((banyakNilai*jumlahNilaiTugasKuadrat)-kuadratJumlahNilaiTugas)/(banyakNilai*(banyakNilai-1))

# MAKSIMUM
nilaiMaksimum=max(nilaiTugas)

# MINIMUM
nilaiMinimum=min(nilaiTugas)

# MODUS
kemunculan=[]
for i in nilaiTugas:
    jumlahMuncul = nilaiTugas.count(i)
    nilai = (jumlahMuncul,i)
    if nilai not in kemunculan:
        kemunculan.append(nilai)
kemunculan.sort(reverse=True)

print("\n Banyak Nilai =",banyakNilai,"\n")
print("Nilai Random:\n",nilaiTugas,"\n")
print("1. Nilai Ganjil Ada",len(nilaiGanjil),"Records:\n",nilaiGanjil,"\n")
print("2. Nilai > 70 Ada",len(besarDari70),"Records:\n",besarDari70,"\n")
print("3. Standar Deviasi:")
print("# Nilai Tugas Kuadrat =",nilaiTugasKuadrat)
print("# Jumlah Nilai Tugas =",jumlahNilaiTugas)
print("# Jumlah dari Nilai Tugas Kuadrat =",jumlahNilaiTugasKuadrat)
print("# Kuadrat dari Jumlah Nilai Tugas =",kuadratJumlahNilaiTugas)
print("# Varian = ((Banyak Nilai * Jumlah dari Nilai Tugas Kuadrat) - Kuadrat dari Jumlah Nilai Tugas ) / (Banyak Nilai * (Banyak Nilai - 1))")
print("# Varian = ((",banyakNilai,"*",jumlahNilaiTugasKuadrat,") -",kuadratJumlahNilaiTugas,") / (",banyakNilai,"* (",banyakNilai,"- 1))")
print("# Varian =",varian)
print("# Standar Deviasi = Akar Kuadrat dari Varian")
print("# Standar Deviasi =",pow(varian,0.5),"\n")
print("4. Nilai Maksimum :",nilaiMaksimum,"\n")
print("5. Nilai Minimum :",nilaiMinimum,"\n")
if kemunculan[0][0]>kemunculan[1][0]:
    print("6. Nilai Modus:", kemunculan[0][1], ", Muncul", kemunculan[0][0], "Kali")
else:
    print("6. Nilai Modus: Tidak Tunggal")
print("\n",kemunculan)
```
